Este notebook faz uma **limpeza passo a passo** no dataset **Netflix Movies and TV Shows** do Kaggle (`netflix_titles.csv`) e gera **gráficos** simples para análise.

### 1) Fazendo import e carregando o dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix_titles.csv")
```

### 2) Visualizando a base de dados

```python
print("Base original:")
print(df.head())
print(df.info())
```

#### Problemas identificados

- Muitos valores nulos em "director", "cast" e "country"
- Duplicatas possíveis
- Coluna "date_added" como string
- Campos inconsistentes de país

### 3) Limpando os dados

Essas são as etapas de limpeza:

- 3.1) Remover as duplicadas
- 3.2) Tratar valores nulos
- 3.3) Converter `date_added` para datetime
- 3.4) Padronizar países

#### 3.1) Removendo os valores duplicados

```python
df = df.drop_duplicates()
```

#### 3.2) Tratando os valores nulos

```python
df["director"] = df["director"].fillna("Desconhecido")
df["cast"] = df["cast"].fillna("Não informado")
df["country"] = df["country"].fillna("Indefinido")
```

#### 3.3) Convertendo a coluna `date_added` para datetime

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
```

#### 3.4) Padronizando os países

```python
df["country"] = df["country"].str.strip()
```

### 4) Análise dos dados

Para isso vamos fazer duas análises:

- Filmes vs Séries
- Top 10 países com mais títulos

#### Filmes vs Séries

```python
conteudo = df["type"].value_counts()
```

#### Top 10 países com mais títulos

```python
paises = df["country"].value_counts().head(10)
```

### 5) Criando os gráficos

```python
plt.rcParams["figure.figsize"] = (9, 5)
plt.rcParams["axes.grid"] = True
plt.rcParams["grid.linestyle"] = ":"
plt.rcParams["grid.alpha"] = 0.4

def add_value_labels(ax):
    for p in ax.patches:
        valor = int(p.get_height())
        ax.annotate(
            f"{valor}",
            (p.get_x() + p.get_width() / 2, p.get_height()),
            ha="center", va="bottom",
            xytext=(0, 3), textcoords="offset points"
        )
```

#### Filmes vs Séries

```python
ax1 = conteudo.plot(kind="bar", color=["red", "blue"])
ax1.set_title("Distribuição de Filmes e Séries na Netflix")
ax1.set_xlabel("Tipo")
ax1.set_ylabel("Quantidade")
add_value_labels(ax1)
plt.tight_layout()
plt.show()
```

#### Top 10 países com mais títulos

```python
ax2 = paises.plot(kind="bar", color="green")
ax2.set_title("Top 10 Países com Mais Títulos na Netflix")
ax2.set_xlabel("País")
ax2.set_ylabel("Quantidade")
plt.xticks(rotation=30, ha="right")
add_value_labels(ax2)
plt.tight_layout()
plt.show()
```
